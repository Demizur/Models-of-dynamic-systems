```python

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
s,r,b=10,0,8/3
def f(p0, t):
    x, y, z = p0
    return [s*(y-x),
    -y+(r-z)*x,
    -b*z+x*y]
t = np.linspace(0,20,2001)
P0 = [0, 1, 1]
for r in [0, 10, 20, 24.06,100,200,24.74]:
    [x,y,z]=odeint(f, P0, t, full_output=False).T
    plt.plot(x,z, color='black', linestyle=' ', marker='.',
    markersize=2)
    plt.xlabel('x')
    plt.ylabel('z')
    plt.grid(True)
    plt.title('Проекция траектории Лоренца на плоскость xz при r=%s'%r)
    plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(Y, t):
    y1, y2 = Y
    return [3*y1-3*y2, y1-y2]

y1 = np.linspace(0, 1000, 20)
y2 = np.linspace(4, 1000, 20)

Y1, Y2 = np.meshgrid(y1, y2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
     

Q=plt.quiver(Y1, Y2, u, v, color='r')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```
